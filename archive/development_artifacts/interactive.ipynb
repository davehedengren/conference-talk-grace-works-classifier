```python
#sns lineplot of score by year
sns.set_theme(style="darkgrid")

# Create the line plot
plt.figure(figsize=(12, 6)) # Adjust figure size for better readability
line_plot = sns.lineplot(
    x='year', 
    y='score', 
    data=df,
    marker='o',  # Add markers to data points
    estimator='mean', # Plot the mean score for each year (default)
    errorbar=('ci', 95) # Show 95% confidence interval (default)
)

# Set the title and axis labels
line_plot.set_title('Average Score by Year', fontsize=16)
line_plot.set_xlabel('Year', fontsize=14)
line_plot.set_ylabel('Average Works vs Grace of Conference talks', fontsize=14)

# Optional: Improve x-axis tick readability if there are many years
# plt.xticks(rotation=45, ha='right')

# Ensure the layout is tight
plt.tight_layout()

line_plot.set_ylabel('Average Works vs Grace of Conference Talk', fontsize=14)
plt.show()
```
